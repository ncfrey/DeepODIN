```
df = pd.read_csv(drive_dir + 'Overdose_Information_Network_Data_CY_January_2018_-_Current_Monthly_County_State_Police.csv')
df1 = pd.read_csv(drive_dir + 'total_data.csv')
```

```
df['Month'] = df['Incident Date'].apply(lambda x: x.split('/')[0])
```

```
df1 = df1.drop('County',axis=1)
```

```
df = df[['Month', 'Day', 'Incident County FIPS Code', 'Age Range', 'Naloxone Administered', 'Dose Count', 'Survive']]
```

```
df2 = df.query('Survive != "U"')
```

```
df2['Naloxone Administered'] = pd.Categorical(df2['Naloxone Administered'])
df2['Naloxone Administered'] = df2['Naloxone Administered'].cat.codes
df2['Survive'] = pd.Categorical(df2['Survive'])
df2['Survive'] = df2["Survive"].cat.codes
df2['Day'] = pd.Categorical(df2['Day'])
df2['Day'] = df2["Day"].cat.codes
df2['Month'] = pd.Categorical(df2['Month'])
df2['Month'] = df2["Month"].cat.codes
df2['Age Range'] = pd.Categorical(df2['Age Range'])
df2['Age Range'] = df2["Age Range"].cat.codes

#df2['Survive'] = df['Survive'].map({'Y': 1, 'N': 0})
#df2['Naloxone Administered'] = df['Naloxone Administered'].map({'Y': 1, 'N': 0})
#df2['Day'] = df['Day'].map({'Monday': 0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3, 'Friday': 4,
#                           'Saturday': 5, 'Sunday': 6})
```

```
fips1 = df2.groupby('Incident County FIPS Code')
fips2 = df1.groupby('FIPS')
```

```
total_data = pd.DataFrame([])
n = 0
for i in fips1.groups.keys():
  g = fips1.get_group(i)
  g1 = fips2.get_group(i)
  for j in g1.columns:
    g[j] = g1[j].mean(axis=0)
  if n == 0:
    total_data = g
  else:
    total_data = pd.concat((total_data,g),axis=0)
  n += 1
  print(total_data.shape,'\n')
```

```
total_data.columns
```

```
total_data = total_data.drop(['Unnamed: 0','FIPS'],axis=1)
```

```
total_data.to_json(drive_dir + "overdose_info_update.json")

target = total_data.pop('Survive')
total_data.fillna(value=0, inplace=True)
```

```
total_data = pd.read_json(drive_dir + 'overdose_info_update.json')
#target = total_data.pop('Survive')
#total_data.fillna(value=0, inplace=True)
```

```
t = total_data.query('Survive ==1')
t = t[:3000]
k = total_data.query('Survive == 0')

total_data = pd.concat((t,k),axis=0)
print(len(total_data))
```

```
target = total_data.pop('Survive')
total_data.fillna(value=0, inplace=True)
```

```
dataset = tf.data.Dataset.from_tensor_slices((total_data.values, target.values))
```

```
dataset = dataset.shuffle(len(total_data))
```

```
train_size = int(0.7 * len(total_data))
val_size = int(0.3 * len(total_data))

train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size)
val_dataset = val_dataset.take(val_size)

train_dataset = train_dataset.batch(32)
val_dataset = val_dataset.batch(32)
```

```
total_data.shape
```

```
from tensorflow import keras
```

```
from tensorflow.keras import layers

#del model


model = tf.keras.Sequential()
# Adds a densely-connected layer with units to the model:
model.add(layers.Dense(30, input_dim=17, activation='relu')) 
#model.add(layers.BatchNormalization())
#model.add(layers.Dropout(0.2))
model.add(layers.Dense(200, input_dim=17,activation='relu'))
#model.add(layers.BatchNormalization())
#model.add(layers.Dropout(0.2))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(80, activation='relu'))
model.add(layers.Dense(50, activation='relu'))
model.add(layers.Dense(30, activation='relu'))
#model.add(layers.BatchNormalization())
model.add(layers.Dense(10,activation='relu'))
# 6 input features
# Add a layer with output units:
model.add(layers.Dense(1, activation='sigmoid'))
```

```
optimizer = keras.optimizers.SGD()
# These depend on the optimizer class
#optimizer_state = [optimizer.iterations, optimizer.lr, optimizer.beta_1,
#                   optimizer.beta_2, optimizer.decay]
#optimizer_reset = tf.variables_initializer(optimizer_state)
#
model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
import datetime

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

# with tf.device('/device:GPU:0'):
model.fit(train_dataset, epochs=100,validation_data=val_dataset)#,callbacks=[tensorboard_callback])
```
